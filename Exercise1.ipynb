```python
plt.figure(figsize=(10,10))
plt.title("Plots")
plt.xlim([-10, 10])
plt.ylim([-10, 10])
x = range(-100,101)
y = []
for i in range(-100,101):
    y.append(i)
plt.plot(x,y,'g--',label='y = x')

x = range(-100,101)
y = []
for i in range(-100,101):
    y.append(i**2)
plt.plot(x,y,'b--',label='y = x^2')

import numpy as np
x = list(range(-100,101))
y = []
for i in range(-100,101):
    m = -2*i
    y.append(np.exp(m))
plt.plot(x,y,'y--',label='y = e^-2x')

x = list(range(-100,101))
y = []
for i in range(-100,101):
    m = np.abs(i)
    y.append(np.exp(-m))
plt.plot(x,y,'r--',label='y = e^-|x|')

x = list(range(0,11))
y = []
for i in range(0,11):
    m = -i
    y.append(i*np.exp(-m))
plt.plot(x,y,'c--',label='y = xe^-x')

x = list(range(-100,101))
y = []
for i in range(-100,101):
    m = i**2
    y.append(np.exp(-m))
plt.plot(x,y,'m--',label='y = e^-x^2')

x = list(range(-100,101))
y = []
for i in range(-100,101):
    m = i
    y.append(np.exp(-m))
plt.plot(x,y,'--',color='#CD2626',label='y = e^-x')

x = list(range(-100,101))
y = []
for i in range(-100,101):
    y.append(i**2+12*i+10)
plt.plot(x,y,'--',color='#8B795E',label='y = x^2 + 12x + 10')
plt.legend(loc='upper left')
plt.grid()
```

#### Gradient Descent

```python
import numpy as np
plt.figure(figsize=(7,7))
plt.xlim([-5,5])
plt.ylim([-20,20])
x = np.linspace(-5,5,200)
y = []
y_bar = []
for i in x:
    y.append(i**2+2*i+3)
plt.plot(x,y)
x_min = y[0]
grad_y = 2*x_min + 2
gy = []
y_n = []
for i in range(200):
    x_min = x_min - 0.1*grad_y
    grad_y = 2*x_min + 2
    gy.append(x_min)
    yn = x_min**2+2*x_min+3
    y_n.append(yn)

plt.plot(gy,y_n,'o--')
plt.grid()
plt.annotate('minimum',(x_min,yn),xytext=(3, 1.5),arrowprops=dict(arrowstyle="->", connectionstyle="arc3"))
print("Minimum x value:",x_min)
print("MInimum value of the function:",x_min**2 + 2*x_min+3)
```

```python
# Problem Set 2
import numpy as np

plt.figure(figsize=(8,8))
plt.title("Problem Set 2 - Activation Functions and their derivatives")
plt.xlim([-3, 3])
plt.ylim([-1.5, 1.5])
plt.xlabel('x')
plt.ylabel('y')
x = np.linspace(-3,3,1000)
y = []
for i in x:
    y.append(1/(1 + np.exp(-i)))
plt.plot(x,y,'g--',label='sigmoid')

y_grad = np.gradient(y,x)

plt.plot(x,y_grad,'b--',label='sigmoid derivative')

y= []

x = np.linspace(-3,3,1000)

for i in x:
    y.append((np.exp(i)-np.exp(-i))/(np.exp(i)+np.exp(-i)))
plt.plot(x,y,'r--',label='tanh')

y_grad = np.gradient(y,x)

plt.plot(x,y_grad,'y--',label='tanh derivative')

y=[]
x = np.linspace(-3,3,1000)
for i in x:
    if i>0:
        y.append(i)
    else:
        y.append(0)
plt.plot(x,y,'m--',label='relu')

y_grad = np.gradient(y,x)

plt.plot(x,y_grad,'c--',label='relu derivative')
y=[]
x = np.linspace(-3,3,1000)
for i in x:
    if i>0:
        y.append(i)
    else: 
        y.append(0.01*i)
plt.plot(x,y,'--',color='navy',label='leaky relu')

y_grad = np.gradient(y,x)

plt.plot(x,y_grad,'--',color='#8B795E',label='leaky relu derivative')

y=[]
x = np.linspace(-3,3,1000)
for i in x:
    if i>=0:
        y.append(i)
    else: 
        y.append(1.0*(np.exp(i)-1))
plt.plot(x,y,'--',color='orange',label='elu')

y_grad = np.gradient(y,x)

plt.plot(x,y_grad,'--',color='purple',label='elu derivative')
plt.grid()
plt.legend(loc='upper left')
```

#### Problem Set 4: Piecewise functions

```python
#Problem set - 4
plt.figure(figsize=(8,8))
plt.title("Problem Set 4: Piecewise functions")
plt.xlim([-3, 3])
plt.ylim([-3, 3])

x = np.linspace(-3,3,1000)
y = []
for i in x:
    if i<0:
        y.append(i*i)
    else:
        y.append(2*i)
plt.plot(x,y,'--',color='orange',label='Plot 1')


x = np.linspace(-3,3,1000)
y = []
for i in x:
    if i<-1:
        y.append(i*i*i)
    elif i>=-1 and i<2:
        m = i*i
        y.append(-m)
    else:
        y.append(i)
plt.plot(x,y,'--',color='blue',label='Plot 2')
plt.grid()
plt.legend(loc='upper left')

```
